```python
with open('train.loss','r') as txt:
    train1 = txt.read()

train1 = train1.split('\n')
train1.pop() #removing the last space in the list
train1 = [float(item.replace(' ','').split(':')[1]) for item in train1]
xaxis1 = []
for i in range(0,len(train1)):
    xaxis1.append(float(i*2500))
```

```python
with open('train2.loss','r') as txt:
    train2 = txt.read()

train2 = train2.split('\n')
train2.pop() #removing the last space in the list
train2 = [float(item.replace(' ','').split(':')[1]) for item in train2]
xaxis2 = []
for i in range(0,len(train2)):
    xaxis2.append(float(i*2500))
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6), dpi=80)
plt.plot(xaxis1,train1,'-',label="beam size 1")
plt.plot(xaxis2,train2,'-',label="beam size 2")
plt.legend(loc='lower right')
plt.xlabel("Validations at training intervals")
plt.ylabel("CIDEr Score")
plt.title("CIDEr validation score")
plt.show()
```
